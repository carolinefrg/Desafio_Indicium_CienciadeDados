#3) Explique como você faria a previsão da nota do imdb a partir dos dados. Quais variáveis e/ou suas transformações você utilizou e por quê? Qual tipo de problema estamos resolvendo (regressão, classificação)? Qual modelo melhor se aproxima dos dados e quais seus prós e contras? Qual medida de performance do modelo foi escolhida e por quê?

```python
# Escolhendo a variável de predição IMDB_Rating
y = filmes_data.IMDB_Rating
```

```python
# Escolhendo as features
filmes_data_features = ['Genre','Runtime','No_of_Votes','Certificate','Gross']
```

```python
# Transformando algumas features categóricas em númericas para o modelo conseguir trabalhar com elas
filmes_data['Certificate'] = pd.Categorical(filmes_data['Certificate']).codes
filmes_data['Genre'] = pd.Categorical(filmes_data['Genre']).codes
filmes_data['Runtime'] = pd.Categorical(filmes_data['Runtime']).codes
```

```python
# Charameremos o conjunto das features citadas anteriormente de X
X = filmes_data[filmes_data_features]
```

```python
X.head()
```

```python
# Escolhendo o modelo preditivo Árvode de Decisão
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

filmes_data_model = DecisionTreeRegressor()

# 20% dos dados serão de teste e os outros 80% serão de treino
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state = 1)

# Ajustando o modelo com os dados de treino
filmes_data_model.fit(train_X,train_y)

# Predição dos valores
val_predicoes = filmes_data_model.predict(val_X)
print(val_predicoes)
```

```python
# Métricas de erro
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

# Erro médio absoluto (MAE)
val_mae = mean_absolute_error(val_y, val_predicoes)
print(f"MAE: {val_mae:.2}")

# Erro Quadrático Médio (MSE)
val_mse = mean_squared_error(val_y, val_predicoes)
print(f"MSE: {val_mse:.2}")

# Erro Percentual Médio Absoluto (MAPE)
def mean_absolute_percentage_error(val_y, val_predicoes):
    val_y, val_predicoes = np.array(val_y), np.array(val_predicoes)
    return np.mean(np.abs((val_y - val_predicoes) / val_y)) * 100

val_mape = mean_absolute_percentage_error(val_y, val_predicoes)
print(f"MAPE: {val_mape:.2}%")
```

```python
# Escolhendo o modelo Random Forest
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()

# 20% dos dados serão dados de teste e os outros 80% serão de treino
treino_X, valor_X, treino_y, valor_y = train_test_split(X, y, test_size=0.2, random_state = 1)

# Ajustando o modelo com os dados de treino
forest_model.fit(treino_X, treino_y)

# Predição dos valores
valor_predicoes = forest_model.predict(valor_X)
print(valor_predicoes)
```

```python
# Erro médio absoluto (MAE)
valor_mae = mean_absolute_error(valor_y, valor_predicoes)
print(f"MAE: {valor_mae:.2}")

# Erro Quadrático Médio (MSE)
valor_mse = mean_squared_error(valor_y, valor_predicoes)
print(f"MSE: {valor_mse:.2}")

# Erro Percentual Médio Absoluto (MAPE)
def mean_absolute_percentage_error(valor_y, valor_predicoes):
    valor_y, valor_predicoes = np.array(valor_y), np.array(valor_predicoes)
    return np.mean(np.abs((valor_y - valor_predicoes) / valor_y)) * 100

valor_mape = mean_absolute_percentage_error(valor_y, valor_predicoes)
print(f"MAPE: {valor_mape:.2}%")
```

#### Seguiria os seguintes passos para fazer a previsão de notas do IMDB:
- Selecionaria as features que contivessem o gênero, o tempo de duração, o número de votos, a classificação e o faturamento (colunas 'Genre', 'Runtime', 'No_of_Votes',	'Certificate' e	'Gross'). Selecionaria estas colunas pois estão bastante interligadas; filmes de determinados gêneros e classificações e cujas durações assemelham-se representam grandes sucessos, e o contrário também é válido.
- Transformaria as variáveis categóricas de gênero, tempo de duração e classificação em variáveis numéricas, para que possam ser utilizadas adequadamente pelo modelo.
####Estamos resolvendo um problema de regressão, pois queremos prever um valor numérico contínuo (neste caso, a nota do IMDB).
####Para esta previsão, escolhemos utilizar dois modelos de regressão: árvore de decisão e random forest, pois são capazes de captar interações entre as variáveis e relações não lineares.
- Prós da árvore de decisão: fácil interpretabilidade e tempo menor de treinamento para conjuntos de dados pequenos.
- Contras da árvore de decisão: sensibilidade à pequenas mudanças nos dados de treinamento e propensão ao overfitting (especialmente em conjuntos grandes e complexos de dados).
- Prós do random forest: menor overfitting em relação à árvore de decisão e boa desenvoltura com conjuntos de dados grandes e complexos.
- Contras do random forest: menores interpretabilidade e tempo de treinamento em relação ao modelo da árvore de decisão.
####Foram escolhidas as medidas de performance MAE, MSE e MAPE. Escolhemos MAE e MSE para termos um contraponto, pois enquanto MAE é mais facilmente interpretável e menos sensível a grandes erros, MSE é o contrário: menos intuitiva e mais sensível a erros. MAPE foi escolhida pois é intuitiva, já que expressa os erros em porcentagem.
